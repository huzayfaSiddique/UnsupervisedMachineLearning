```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
df=scalar.fit_transform(df)
```

```python
df
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca
```

```python
pca_scaled=pca.fit_transform(df)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,15))
plt.title("DENDROGRAM")
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.xlabel("Sample Index")
plt.ylabel("Eucledian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil=[]
for k in range(2,11):
  agg=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward')
  agg.fit(pca_scaled)
  score=silhouette_score(pca_scaled,agg.labels_)
  sil.append(score)
```

```python
plt.plot(range(2,11),sil)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("silhoutte coefficients")
```

### ***Hierarchical*** ***Clustering***

Hierarchical clustering is a clustering method which builds a hierarchy of clusters step by step. Instead of fixing K at the start like K-means, it either keeps merging smaller clusters into bigger ones (agglomerative) or keeps splitting one big cluster into smaller ones (divisive).But the main difference is that in hierarchical clustering we donâ€™t use centroids at all.

#### Hyperparameters:

The hyperparameters are basically:

1.The linkage criteria (single linkage, complete linkage, average linkage etc).

2.The distance metric (like Euclidean distance, Manhattan distance, etc).

3.The number of clusters K (which we later decide from dendrogram).

#### Types of Hierarchical Clustering

There are two types:

1.Agglomerative (bottom up, most common)

2.Divisive (top down, not used so much)

#### Steps of Agglomerative

1.Start with each point as its own cluster.

2.Find the nearest neighboring clusters (based on distance metric + linkage you select).

3.Merge them to form a new cluster.

4,Repeat the process until everything becomes one big cluster.

#### Steps of Divisive

Divisive is just opposite: start with all points in one cluster and then keep splitting until each point is separate.

#### How to decide distance between clusters:

Here comes the question: if we have points near and far, how do we decide? There are different linkage methods:

1.Single linkage: distance between the closest two points of the clusters.

2.Complete linkage: distance between the farthest two points of the clusters.

3.Average linkage: distance between the centroids or average of all pairwise distances.

#### How to choose number of clusters (K):

For hierarchical clustering, we use a dendrogram.

Dendrogram is formed by calculating Euclidean distance (usually between points, not clusters at start, then clusters when they are merged).
We draw vertical lines for merges and connect them with horizontal lines showing distance at which merge happened.

#### To decide K:

Look for the longest vertical line in the dendrogram that is not crossed by any horizontal line.
The number of vertical cuts we can make gives us the number of clusters.

This is known as the elbow-like method for dendrogram.
